Run_ID - The identification of each machining run

Exp - the preparation phase

Replica - The replica's number

Tool_ID - The identification of the tool used

Group - General information

Subgroup - 	General information

Position - The position where roughness was measured

Condition	- The condition related to the experiment's design.

TCond	- The flank wear width (VB) measured after machining. [mm]

Machined_length	- The machined length [mm]

Init_diameter	- The diameter before machining [mm]

Final_diameter	- The diameter after machining [mm]

CTime	- The cutting time [min]

R_measurement	- The angular position (on the workpiece) of roughness measurement.

ap	- Depth of cut [mm]

vc	- Cutting speed [m/min]

f	- Feed rate [mm/rev]

Ra	- Average mean deviation [um]

Rz	- Maximum height [um]

Rsk	- Skewness

Rku	- Kurtosis

RSm	- Mean width of profile [um]

Rt - Total height of profile [um]

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X1 = df1[[ 'Machined_length', 'Init_diameter', 'Final_diameter', 'CTime', 'R_measurement', 'ap', 'vc', 'f', 'Ra', 'Rz', 'Rsk', 'Rku', 'RSm','Rt']]
y1 = df1['TCond']
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.2, random_state=42)

# Linear Regression Model
linear_model = LinearRegression()
linear_model.fit(X1_train, y1_train)
linear_y1_pred = linear_model.predict(X1_test)
linear_mse1 = mean_squared_error(y1_test, linear_y1_pred)
linear_rmse1 = np.sqrt(linear_mse1)
linear_r21 = r2_score(y1_test, linear_y1_pred)
print('Linear Regression Mean squared error: ', linear_mse1)
print('Linear Regression Root Mean Squared Error:', linear_rmse1)
print('Linear Regression R-squared: ', linear_r21)

# Decision Tree Regression Model
tree_model = DecisionTreeRegressor(max_depth=3)
tree_model.fit(X1_train, y1_train)
tree_y1_pred = tree_model.predict(X1_test)
tree_mse1 = mean_squared_error(y1_test, tree_y1_pred)
tree_rmse1 = np.sqrt(tree_mse1)
tree_r21 = r2_score(y1_test, tree_y1_pred)
print('Decision Tree Regression Mean squared error: ', tree_mse1)
print('Decision Tree Regression Root Mean Squared Error:', tree_rmse1)
print('Decision Tree Regression R-squared: ', tree_r21)

# Random Forest Regression Model
forest_model = RandomForestRegressor(n_estimators=10)
forest_model.fit(X1_train, y1_train)
forest_y1_pred = forest_model.predict(X1_test)
forest_mse1 = mean_squared_error(y1_test, forest_y1_pred)
forest_rmse1 = np.sqrt(forest_mse1)
forest_r21 = r2_score(y1_test, forest_y1_pred)
print('Random Forest Regression Mean squared error: ', forest_mse1)
print('Random Forest Regression Root Mean Squared Error:', forest_rmse1)
print('Random Forest Regression R-squared: ', forest_r21)
```
